```python
import os.path
import sys
import chigger
from IPython.display import Image
import vtk
```

```python
camera = vtk.vtkCamera()
camera.SetViewUp(0.0000, 1.0000, 0.0000)
camera.SetPosition(-5,-5,12)
camera.SetFocalPoint(0,0,0)

file_name = os.path.join('..', '..', 'tests/input/mug_blocks_out.e')
reader = chigger.exodus.ExodusReader(file_name, adaptive=False)
result = chigger.exodus.ExodusResult(reader, cmap='viridis', 
                                     edge_color=[0,0,0], range=[0, 1], camera=camera)
window = chigger.RenderWindow(result, size=[300,300], test=True, offscreen=True)
Image(window.getImageData())
```
